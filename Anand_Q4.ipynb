```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)

# Generate data for topics and difficulty levels
topics = ["Calculus", "Linear Algebra", "Probability & Statistics", "Differential Equations", "Discrete Mathematics",
          "Abstract Algebra", "Real Analysis", "Topology"]

difficulty_levels = ["Beginner", "Intermediate", "Advanced"]

countries = ["India", "United States", "China", "Germany", "United Kingdom", "Canada", "Australia"]

# Simulated data: Number of problems solved by topic
topic_data = {topic: np.random.randint(500, 1500) for topic in topics}
difficulty_data = {level: np.random.randint(1000, 3000) for level in difficulty_levels}
country_data = {country: np.random.randint(2000, 10000) for country in countries}

# Convert to DataFrames
topics_df = pd.DataFrame(list(topic_data.items()), columns=["Topic", "Problems Solved"])
difficulty_df = pd.DataFrame(list(difficulty_data.items()), columns=["Difficulty Level", "Problems Solved"])
country_df = pd.DataFrame(list(country_data.items()), columns=["Country", "Problems Solved"])

# Display Results
print("Most Difficult Topics:")
print(topics_df.sort_values(by="Problems Solved", ascending=False), "\n")

print("Questions Solved by Difficulty Level:")
print(difficulty_df.sort_values(by="Problems Solved", ascending=False), "\n")

print("Math Problems Solved by Country:")
print(country_df.sort_values(by="Problems Solved", ascending=False), "\n")

# Plot data
plt.figure(figsize=(10, 6))

# Plot for Topics
plt.subplot(1, 3, 1)
plt.barh(topics_df["Topic"], topics_df["Problems Solved"], color="skyblue")
plt.xlabel("Problems Solved")
plt.ylabel("Topics")
plt.title("Problems Solved by Topic")

# Plot for Difficulty
plt.subplot(1, 3, 2)
plt.bar(difficulty_df["Difficulty Level"], difficulty_df["Problems Solved"], color="salmon")
plt.xlabel("Difficulty Level")
plt.ylabel("Problems Solved")
plt.title("Problems Solved by Difficulty Level")

# Plot for Country
plt.subplot(1, 3, 3)
plt.bar(country_df["Country"], country_df["Problems Solved"], color="lightgreen")
plt.xlabel("Countries")
plt.ylabel("Problems Solved")
plt.title("Problems Solved by Country")

plt.tight_layout()
plt.show()
```
