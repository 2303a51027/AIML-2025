```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('/content/sample_data/turkishCF.csv', sep=';', on_bad_lines='skip')

# Rename columns for easier access
data.rename(columns={
    'proje_adi': 'Startup Name',
    'toplanan_tutar': 'Crowdsourced Funding',
    'proje_sahibi_cinsiyet': 'Gender of Founder',
    'kategori': 'Technology',
    'platform_adi': 'Platform'
}, inplace=True)

# Convert relevant columns to numeric
data['Crowdsourced Funding'] = pd.to_numeric(data['Crowdsourced Funding'], errors='coerce')

# Q1: Top-10 Startups by Crowdsourced Funding
top_10_startups = data.sort_values(by='Crowdsourced Funding', ascending=False).head(10)
print("Top-10 Startups in Turkey:")
print(top_10_startups[['Startup Name', 'Crowdsourced Funding']])

# Q2: Max and Min Crowdsourced Funding
max_funding = data['Crowdsourced Funding'].max()
min_funding = data['Crowdsourced Funding'].min()
print(f"Maximum Crowdsourced Funding: {max_funding}")
print(f"Minimum Crowdsourced Funding: {min_funding}")

# Q3: Gender with the Most Funding
gender_funding = data.groupby('Gender of Founder')['Crowdsourced Funding'].sum()
top_gender = gender_funding.idxmax()
top_gender_funding = gender_funding.max()
print(f"Gender with the most funding: {top_gender} with {top_gender_funding} in funding")

# Q4: Top-5 Cities with Most Startups
city_count = data['Platform'].value_counts().head(5)
print("Top-5 Cities with Most Startups:")
print(city_count)

# Q5: Top-5 Technologies in Turkish Startups
technology_count = data['Technology'].value_counts().head(5)
print("Top-5 Technologies Popular in Turkish Startups:")
print(technology_count)

# Visualization: Top-5 Cities with Most Startups
plt.figure(figsize=(10, 6))
sns.barplot(x=city_count.index, y=city_count.values, palette='viridis')
plt.title("Top-5 Cities with Most Startups")
plt.xlabel("City")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.show()

# Visualization: Total Crowdsourced Funding by Gender
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_funding.index, y=gender_funding.values, palette='magma')
plt.title("Total Crowdsourced Funding by Gender")
plt.xlabel("Gender of Founder")
plt.ylabel("Total Crowdsourced Funding")
plt.show()

# Visualization: Top-5 Popular Technologies in Turkish Startups
plt.figure(figsize=(10, 6))
sns.barplot(x=technology_count.index, y=technology_count.values, palette='coolwarm')
plt.title("Top-5 Popular Technologies in Turkish Startups")
plt.xlabel("Technology")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.show()
```
